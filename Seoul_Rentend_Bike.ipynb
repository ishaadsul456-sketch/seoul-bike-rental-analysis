Data Visualization

```python
#Impact of Weather Conditions on Bike Rentals
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Temperature vs Rentals
df.groupby('Temperature(°C)')['Rented Bike Count'].mean().plot(kind='line', marker='o', color='tomato', ax=axs[0])
axs[0].set_title('Avg Rentals vs. Temperature')
axs[0].set_xlabel('Temperature (°C)')
axs[0].set_ylabel('Average Rentals')
axs[0].grid(True, linestyle='--', alpha=0.4)

# Visibility vs Rentals
df.groupby('Visibility (10m)')['Rented Bike Count'].mean().plot(kind='line', marker='o', color='royalblue', ax=axs[1])
axs[1].set_title('Avg Rentals vs. Visibility')
axs[1].set_xlabel('Visibility (10m)')
axs[1].set_ylabel('')
axs[1].grid(True, linestyle='--', alpha=0.4)

# Humidity vs Rentals
df.groupby('Humidity(%)')['Rented Bike Count'].mean().plot(kind='line', marker='o', color='forestgreen', ax=axs[2])
axs[2].set_title('Avg Rentals vs. Humidity')
axs[2].set_xlabel('Humidity (%)')
axs[2].set_ylabel('')
axs[2].grid(True, linestyle='--', alpha=0.4)

# Overall layout
plt.suptitle('Impact of Weather Conditions on Bike Rentals', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python


# Step 1: Create bins for Dew Point Temperature
df['Dew Point Bin'] = pd.cut(df['Dew point temperature(°C)'],
                                   bins=[-30, -10, 0, 10, 20, 30],
                                   labels=['< -10°C', '-10°C to 0°C', '0°C to 10°C', '10°C to 20°C', '20°C to 30°C'])

# Step 2: Group by the bins and calculate average rentals
avg_rentals_by_dew = df.groupby('Dew Point Bin')['Rented Bike Count'].mean()

# Step 3: Plot
avg_rentals_by_dew.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Rentals by Dew Point Temperature')
plt.xlabel('Dew Point Temperature Range')
plt.ylabel('Average Bike Rentals')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Does solar radiation affect bike rentals?
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Solar Radiation (MJ/m2)', y='Rented Bike Count', color='orange', alpha=0.6)

plt.title('Effect of Solar Radiation on Bike Rentals')
plt.xlabel('Solar Radiation (MJ/m²)')
plt.ylabel('Rented Bike Count')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#3.Which season has the highest average rentals?
sns.barplot(data=df, x='Seasons', y='Rented Bike Count', estimator='mean')
plt.title('Average Bike Rentals by Season')
```

```python
4.#Does rainfall or snowfall affect bike rentals?


df['Rainfall Category'] = df['Rainfall(mm)'].apply(lambda x: 'No Rain' if x == 0 else 'Rain')
df['Snowfall Category'] = df['Snowfall (cm)'].apply(lambda x: 'No Snow' if x == 0 else 'Snow')
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(data=df, x='Rainfall Category', y='Rented Bike Count', ax=axes[0])
axes[0].set_title('Effect of Rainfall on Bike Rentals')
axes[0].set_xlabel('Rainfall Category')
axes[0].set_ylabel('Rented Bike Count')

sns.boxplot(data=df, x='Snowfall Category', y='Rented Bike Count', ax=axes[1])
axes[1].set_title('Effect of Snowfall on Bike Rentals')
axes[1].set_xlabel('Snowfall Category')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# 5. What is the distribution of seasons in the data?
df['Seasons'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6),colors=['skyblue','lightblue','darkblue','blue'])
```

```python
#6 Do different seasons + functioning days affect rentals?

# Step 1: Create a pivot table
pivot = df.pivot_table(index='Seasons',
                       columns='Functioning Day',
                       values='Rented Bike Count',
                       aggfunc='mean')

# Step 2: Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(pivot, annot=True, cmap='YlOrRd', fmt='.0f', linewidths=0.5)

plt.title('Average Rentals by Season and Functioning Day')
plt.xlabel('Functioning Day')
plt.ylabel('Season')
plt.tight_layout()
plt.show()

```

```python
#8.Are rentals higher during the day or night?
df['Time of Day'] = df['Hour'].apply(lambda h: 'Day' if 6 <= h <= 20 else 'Night')
time_of_day_rentals = df.groupby('Time of Day')['Rented Bike Count'].sum()
time_of_day_rentals.plot(kind='bar', color=['orange', 'navy'])
plt.title('Total Bike Rentals: Day vs Night')
plt.ylabel('Total Rentals')
plt.xlabel('Time of Day')
plt.tight_layout()
plt.show()
```

Conclusion

The analysis of the bike rental data reveals clear patterns in user behavior based on seasons, time of day, and weather conditions. Rentals are highest during the summer and autumn seasons, with winter showing the lowest usage. Most bikes are rented between 8 AM and 8 PM, especially during typical commuting hours. Rentals are significantly higher on functioning days compared to non-functioning days, while holidays generally see a slight drop in activity. Weather plays a crucial role—temperature and solar radiation have a positive correlation with rentals, suggesting that people prefer riding in warm, sunny conditions. In contrast, humidity, rainfall, and wind speed tend to lower rental counts. Scatter plots effectively visualized the impact of solar radiation and other weather variables, while heatmaps provided insight into how different combinations of season, hour, holiday, and functioning days affect demand. These findings are useful for improving bike availability and planning ahead for seasonal or weather-related demand changes. Overall, the data highlights how environmental and calendar factors influence urban bike-sharing trends.
